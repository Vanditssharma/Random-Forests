# Random Forest Project 

Lending Club connects people who need money (borrowers) with people who have money (investors). Hopefully, as an investor you would want to invest in people who showed a profile of having a high probability of paying you back. Model that will help predict this.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# trend between FICO score and interest rate
sns.jointplot(x='fico',y='int.rate',data=loans,color='purple')
```

```python
#trend differed between not.fully.paid and credit.policy
plt.figure(figsize=(11,7))
sns.lmplot(y='int.rate',x='fico',data=loans,hue='credit.policy',
           col='not.fully.paid',palette='Set1')
```

```python
cat_feats=['purpose']
```

```python
#get dummies
final_data = pd.get_dummies(loans,columns=cat_feats,drop_first=True)
```

```python
final_data.head()
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X= final_data.drop('not.fully.paid', axis=1)
y=final_data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=600)
```

```python
rfc.fit(X_train,y_train)
```

```python
predictions = rfc.predict(X_test)
```

```python
#Evaluating Model
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
